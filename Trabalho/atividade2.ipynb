```python
Q1 = sd['max_playtime'].quantile(0.25)
Q3 = sd['max_playtime'].quantile(0.75)
QTD=Q3-Q1
print(QTD)
```

```python
limite_inferior=Q1-1.5*QTD
limite_superior=Q3+1.5*QTD
Outliers=sd[(sd['max_playtime'] < limite_inferior) | (sd['max_playtime'] > limite_superior)]
print(Outliers)
```

```python
sd['avg_rating'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Column 2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
## Assimetria positiva
```

```python
import numpy as np
```

```python
print(np.log(sd['max_playtime']))
sd = sd[sd['max_playtime'] > 0]
sd['Log']=np.log(sd['max_playtime'])
```

```python
plt.hist(sd['Log'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma dos valores log-transformados')
plt.xlabel('Log(valor)')
plt.ylabel('Frequência')
plt.grid(True)
plt.show()
##exibe assimetria positiva
```

```python
plt.scatter(sd['min_players'], sd['max_players'], alpha=0.7, color='blue', edgecolors='k')

plt.title('Dispersão: min_players vs max_players')
plt.xlabel('Mínimo de Jogadores')
plt.ylabel('Máximo de Jogadores')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
correlacao = sd.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Matriz de Correlação')
plt.show()
```

```python
sd['decada'] = (sd['release_year'] // 10 * 10).astype(int)


jogos_por_decada =sd.groupby('decada')['boardgame'].count().reset_index()
jogos_por_decada = jogos_por_decada.rename(columns={'boardgame': 'jogos'})

print(jogos_por_decada)

## a decada com mais lançamentos foi em 2010
```
