```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações padrão para gráficos (para começar limpo)
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["axes.spines.right"] = False
plt.rcParams["axes.spines.top"] = False

print("Bibliotecas importadas e configurações aplicadas!")
```

```python
# 
np.random.seed(42)
investimento_marketing = np.random.randint(10000, 50000, 12)
vendas_mensais = investimento_marketing * 5 + np.random.randint(-5000, 5000, 12)
df_marketing_vendas = pd.DataFrame({
    'Investimento_Marketing': investimento_marketing,
    'Vendas_Mensais': vendas_mensais
})
print("\nDataFrame para Desafio 1.4 (primeiras 5 linhas):")
print(df_marketing_vendas.head())
```

```python
#
plt.figure(figsize=(8, 5))

sns.regplot(
    x='Investimento_Marketing',
    y='Vendas_Mensais',
    data=df_marketing_vendas
)

plt.title('Relação entre Investimento em Marketing e Vendas')
plt.xlabel('Investimento em Marketing')
plt.ylabel('Vendas Mensais')

plt.show()
```

```python
# Dados para o Desafio 2.1
#O erro estava em iniciar por um valor diferente de zero, ao iniciar por zero se ve que as vendas variaram pouco
dados_trimestrais = {
    'Trimestre': ['Q1', 'Q2', 'Q3', 'Q4'],
    'Vendas': [1000, 1050, 1020, 980]
}
df_trimestral = pd.DataFrame(dados_trimestrais)

plt.figure(figsize=(7, 5))
ax = sns.barplot(x='Trimestre', y='Vendas', data=df_trimestral, color='lightcoral')
ax.set_ylim(bottom=0) # Eixo Y começando alto
ax.set_title('Vendas Trimestrais - Último Ano', fontsize=14, fontweight='bold')
ax.set_xlabel('Trimestre')
ax.set_ylabel('Vendas (Unidades)')
ax.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Dados para o Desafio 2.3
#O erro esta em usar cores sem motivo nenhum, apenas o nome e a diferença de tamanho é o bastante

dados_funcionarios = {
    'Funcionario': ['Ana', 'Bruno', 'Carla', 'Daniel', 'Eva'],
    'Pontuacao': [85, 72, 91, 68, 79]
}
df_funcionarios = pd.DataFrame(dados_funcionarios)

plt.figure(figsize=(7, 5))
ax = sns.barplot(x='Funcionario', y='Pontuacao', data=df_funcionarios)
ax.set_title('Pontuação de Desempenho dos Funcionários', fontsize=14, fontweight='bold')
ax.set_ylim(0, 100)
plt.show()
```

```python
# Dados para o Desafio 2.2
#O erro esta em usar o grafico de pizza com muitas coisas, para isso o de barra é melhor
dados_produtos = {
    'Produto': [f'Produto {i}' for i in range(1, 9)],
    'Vendas': [150, 120, 80, 70, 60, 50, 40, 30]
}
df_produtos = pd.DataFrame(dados_produtos)

plt.figure(figsize=(7, 5))
ax = sns.barplot(x='Produto', y='Vendas', data=df_produtos)
plt.title('Proporção de Vendas por Produto', fontsize=14, fontweight='bold')
ax.set_ylim(0, 160)
plt.xticks(rotation=45)
plt.show()
```

```python
# Dados para o Desafio 2.4
#O erro esta em usar cumsum que distorce o grafico deixando a comparação mais dificil e o uso de pontos e marcaçoes que geram poluição visual, nem como o uso de 'M' inves de 'MS'
datas_mes = pd.date_range(start='2023-01-01', periods=12, freq='MS')
np.random.seed(42)
metrica_a = np.random.randint(50, 150, 12)
metrica_b = np.random.randint(40, 130, 12)

df_metricas = pd.DataFrame({
    'Data': datas_mes,
    'Metrica_A': metrica_a,
    'Metrica_B': metrica_b
})

plt.figure(figsize=(10, 6))
plt.plot(df_metricas['Data'], df_metricas['Metrica_A'], label='Métrica A', linestyle='-', linewidth=2, color='blue')
plt.plot(df_metricas['Data'], df_metricas['Metrica_B'], label='Métrica B', linestyle='-', linewidth=2, color='red')

plt.title('Evolução das Métricas A e B', fontsize=16, fontweight='bold')
plt.xlabel('Data', fontsize=12)
plt.ylabel('Valor', fontsize=12)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.grid(True, linestyle=':', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
dados_funcionalidades = { 'Funcionalidade': [f'Func {i}' for i in range(1, 6)], 'Uso': [0.8, 0.6, 0.9, 0.4, 0.7], 'Satisfacao': [0.9, 0.7, 0.8, 0.5, 0.6], 
                         'Impacto': [100, 50, 120, 30, 80] } 
df_funcionalidades = pd.DataFrame(dados_funcionalidades)

plt.figure(figsize=(8, 8))

scatter = plt.scatter(
    x=df_funcionalidades['Uso'],
    y=df_funcionalidades['Satisfacao'],
    s=df_funcionalidades['Impacto'] * 3,
    alpha=0.7,
    c=df_funcionalidades['Impacto'],
    cmap='viridis'
)

for i, row in df_funcionalidades.iterrows():
    plt.text(row['Uso'], row['Satisfacao'] + 0.02,
             row['Funcionalidade'], fontsize=9, ha='center')

plt.colorbar(scatter, label='Impacto')
plt.title('Desempenho das Funcionalidades', fontsize=14, fontweight='bold')
plt.xlabel('Uso (0-1)')
plt.ylabel('Satisfação (0-1)')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.grid(True, linestyle=':', alpha=0.5)

plt.show()
```
