```python
#Importando o pandas exibir  todas as linhas e resolver o problema com ;
import pandas as pd

pd.set_option('display.max_rows', None)

df = pd.read_csv('atividade3_dataset.csv',sep=";")

print(df.to_string())
```

```python
#Linhas com valores nulos
print(df[df.isnull().any(axis=1)])
```

```python
#Deletar linhas que possuam algum valor nulo
df.dropna(inplace=True)
print(df.to_string())
```

```python
#Tipos 
print(df.info())
```

```python
#Mudar tipos que não fazem sentido e os substituindo, trocando a virgula por ponto para converter
df['Preço'] = df['Preço'].str.replace(',', '.')
df['Preço'] = df['Preço'].astype('float64')
df['Idade'] = df['Idade'].astype('int64')
print(df.dtypes)
```

```python
#Exibi duplicadas 
df[df.duplicated(keep=False)]
print(df)
```

```python
#Contar numero de duplicadas
print(df.duplicated().sum())
```

```python
#Removendo duplicatas
df = df.drop_duplicates()
print(df)
```

```python
#Quanto cada cidade aparece e as cidades com nomes inconsistentes
print(df['Cidade'].value_counts())
```

```python
#Corrigindo o nome das cidades
df['Cidade'] = df['Cidade'].str.strip().str.upper()
print(df['Cidade'].value_counts())
```

```python
#Produto mais escolhido por clientes diferentes
mais_escolhido = df['Produto'].value_counts().idxmax()
print(df['Produto'].value_counts())
```

```python
#Importando para gerar os graficos
import matplotlib.pyplot as plt
```

```python
#Criando uma nova coluna multiplicando a quantidade de um produto pelo preço para saber quanto cada cliente pagou
df['Total'] = df['Preço'] * df['Quantidade']
print(df)
```

```python
#Caculando a media da idade dos clientes para se ter uma ideia de quem é o publico alvo
media=df['Idade'].mean()
print(media)
```

```python
#Grafico dos produtos mais vendidos, escolher melhor a reposição do estoque
produtos_vendidos = df.groupby('Produto')['Quantidade'].sum()
produtos_vendidos = produtos_vendidos.sort_values(ascending=False)
produtos_vendidos.plot(kind='bar', color='skyblue')
plt.title("Produtos Mais Vendidos")
plt.ylabel("Quantidade Vendida")
plt.xlabel("Produto")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Produtos mais vendidos em pizza melhor entendimento
produtos_vendidos.plot(
    kind='pie',
    autopct='%1.1f%%',    # Mostra porcentagens
    startangle=90,        # Começa do topo
    figsize=(6, 6),       # Tamanho da figura
    legend=False          # Oculta legenda (opcional)
)

plt.title("Participação dos Produtos nas Vendas")
plt.ylabel('')  # Remove o rótulo do eixo Y
plt.tight_layout()
plt.show()
```

```python
#Porcentagem de lucro por cidade levando em conta o tanto que ela aparece e quanto dinheiro os clientes de lá gastaram, ideias para expansão de negocios nas cidades
lucro_por_cidade = df.groupby('Cidade')['Total'].sum()
lucro_por_cidade.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    figsize=(7, 7),
    colormap='tab20'
)

plt.title('Participação do Lucro por Cidade (%)')
plt.ylabel('')  # Remove o rótulo do eixo Y
plt.tight_layout()
plt.show()
```
