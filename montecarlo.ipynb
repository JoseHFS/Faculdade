```python
import numpy as np
import matplotlib.pyplot as plt

# Configurações da Simulação
N_SIMULACOES = 10000

# --- PARÂMETROS ESTRATÉGIA A ---
vendas_media_a = 500
vendas_desvio_a = 50
custo_min_a = 600
custo_max_a = 700
preco_venda_a = 1200
investimento_fixo_a = 50000

# --- PARÂMETROS ESTRATÉGIA B ---
vendas_media_b = 800
vendas_desvio_b = 250
custo_min_b = 400
custo_max_b = 800
preco_venda_b = 1000
investimento_fixo_b = 150000

print("Parâmetros carregados com sucesso!")
```

```python
vendas=np.random.normal(500,50,N_SIMULACOES)

custo=np.random.uniform(600, 700, N_SIMULACOES)

lucroa=(vendas*preco_venda_a)-(vendas*custo)-investimento_fixo_a

#print(lucro/N_SIMULACOES)

media_lucroa=np.mean(lucroa)
p5a=(np.percentile(lucroa,5))
p95a=(np.percentile(lucroa,95))
```

```python
vendasb=np.random.normal(800,250,N_SIMULACOES)

custob=np.random.uniform(400, 800, N_SIMULACOES)

lucrob=(vendasb*preco_venda_b)-(vendasb*custob)-investimento_fixo_b

#print(lucro/N_SIMULACOES)

media_lucrob=np.mean(lucrob)
p5b=(np.percentile(lucrob,5))
p95b=(np.percentile(lucrob,95))
```

```python
plt.figure(figsize=(10, 6))
plt.hist(lucrob, bins=50, color='skyblue', edgecolor='black')
plt.hist(lucroa, bins=50, color='red', edgecolor='black')
plt.axvline(media_lucrob, color='red', linestyle='--', label=f'Média: R${media_lucrob:.2f}')
plt.axvline(media_lucroa, color='red', linestyle='--', label=f'Média: R${media_lucroa:.2f}')
plt.axvline(p5b, color='orange', linestyle=':', label='Intervalo de Confiança 90%')
plt.axvline(p5a, color='orange', linestyle=':', label='Intervalo de Confiança 90%')
plt.axvline(p95b, color='orange', linestyle=':')
plt.axvline(p95a, color='orange', linestyle=':')

plt.title('Simulação de Monte Carlo - Projeção de Lucro')
plt.xlabel('Lucro (R$)')
plt.ylabel('Frequência (Quantas vezes aconteceu)')
plt.legend()
plt.show()
```
